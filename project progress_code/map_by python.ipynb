```python
data = pd.read_csv('./map.txt', sep='\t')
data
```

```python
world_map = folium.Map ( location = [10,20], zoom_start = 2.3, tiles = 'Stamen Toner' )
```

```python
for lat, long, confirmed, death, name in zip ( data['Lat'], data['Long'], data['confirmed'], data['death'], data['Country/Region'] ):
    folium.CircleMarker ([lat, long],
                        radius = 10,
                        popup = ('<strong>Country/Region</strong>: ' + str(name).capitalize() + '<br>' 
                                + '<strong>Confirmed</strong>: ' + str(confirmed) + '<br>'
                                + '<strong>Death</strong>: ' + str(death) + '<br>'),
                        color = 'red',
                        fill_color = 'red',
                        fill_opacity = 0.7 ).add_to(world_map)
```
